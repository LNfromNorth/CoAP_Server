```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
thread1_log     =    [13755.2, 13010.7, 13726.3]
thread1_no_log  =    [18622.0, 18477.5, 21132.7]
thread2_log     =    [14164.3, 14749.3, 14351.3]
thread2_no_log  =    [19440.1, 18148.8, 19297.6]
thread3_log     =    [13919.8, 13858.1, 13007.3]
thread3_no_log  =    [19615.5, 18450.2, 19342.4]
thread4_log     =    [11825.9, 13520.8, 12913.2]
thread4_no_log  =    [18982.5, 18867.9, 18782.9]
libcoap         =    [18642.8, 20449.9, 19592.5]
```

```python
def averange(data_list):
    return (40 * (data_list[0] + data_list[1] + data_list[2])/3)/1000
```

```python
x = np.array([1, 2, 3, 4])
y_log = np.array([averange(thread1_log), averange(thread2_log), averange(thread3_log), averange(thread4_log)])
y_no_log = np.array([averange(thread1_no_log), averange(thread2_no_log), averange(thread3_no_log), averange(thread4_no_log)])
```

```python
print(x)
print(y_log)
print(y_no_log)
```

```python
zhfont1 = matplotlib.font_manager.FontProperties(fname="SourceHanSansSC-Bold.otf")

# plt.title("性能测试", fontproperties=zhfont1)
plt.plot(x, y_log, '.-', label='log') #, fontproperties=zhfont1)
plt.plot(x, y_no_log, '.-', label='no log') #, fontproperties=zhfont1)
plt.xticks(x)
plt.xlabel('线程数(个)', fontproperties=zhfont1)
plt.ylabel('吞吐量(KB/s)', fontproperties=zhfont1)
plt.legend()
plt.show()
```
